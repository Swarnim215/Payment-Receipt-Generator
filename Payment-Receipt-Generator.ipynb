```python
from datetime import datetime
```

```python
def receipt_gen():
    

    receipt_no = datetime.now().strftime("%Y%m%dXX%H%M%S")
    date_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    cust_name = input("Enter Customer Name: ")

    item  = []
    t = 0
    
    while True:
        item_name = input("\nEnter the item name: ")
        qty = int(input("Enter Quantity"))
        try:
            rate = eval(input("Enter item amount [IN Rs.]: "))
        except ValueError:
            print("Enter a number as rate")
        
            
        i_total = qty * rate
        t += i_total #t = sub total
        item.append((item_name, qty, rate, i_total)) # 1st() as a function call and 2nd () as we create tuple of the items 
        choice = input("Add another item? [y/n]: ").lower()
        if choice != "y":
            break
        
    payment_mode = input(("Payment mode(UPI/ Cash/ Card): "))
    taxr = 5
    tax = (taxr / 100) * t
    total_amt = t + tax

    print('\n------------- PAYMENT RECEIPT -------------')
    print(f"Receipt No.:        {receipt_no}")
    print(f"Date & Time :        {date_time}")
    print(f"Customer Name:       {cust_name}")
    print("\n---------------------------------------------")
    print(f"Item       Qty       Price       Total")

    for i in item:
        print(f"{i[0]:15} {i[1]:<7} Rs.{i[2]:<7} Rs.{i[3]:.2f}") # '<7' adjusts the 7 aligned space before the previous item & 2f adds 2 decimal floating point

    print("\n---------------------------------------------")
    print(f"Subtotal:             Rs.{t:.2f} ") #:.2f and not .2f only
    print(f"GST(5%):              Rs.{tax:.2f}")
    print(f"Total Payment:        Rs.{total_amt:.2f}")
    print(f"Payment mode {payment_mode}")
    print("PAYMENT SUCCESSFUL")
    print("Thank you for shopping with us!")
    print("\n---------------------------------------------\n")

#SAVING RECEIPT

    filenm = f"receipt_{receipt_no}.txt"
    with open(filenm, "w") as f:
        f.write("PAYMENT RECEIPT\n")
        f.write(f"Receipt No.:         {receipt_no}")
        f.write(f"Date & Time :        {date_time}")
        f.write(f"Customer Name:       {cust_name}")
        f.write("\n---------------------------------------------\n")
        f.write(f"Item       Qty       Price       Total")

        for i in item:
            f.write(f'i[0]')
            # '<7' adjusts the 7 aligned space before the previous item & 2f adds 2 decimal floating point
        # this formatting only works in string, integer and float but doesn't work with TUPLE 

        f.write("\n---------------------------------------------\n")
        f.write(f"Subtotal:             Rs.{t:.2f} ") #:.2f and not .2f only
        f.write(f"GST(5%):              Rs.{tax:.2f}")
        f.write(f"Total Payment:        Rs.{total_amt:.2f}")
        f.write("\nPAYMENT SUCCESSFUL !! ")
        f.write("\nThank you for shopping with us !!")
        f.write("\n---------------------------------------------")

    print(f"\n Receipt {receipt_no} saved!")
            

receipt_gen()
    
```
